__5*.Написать на Python программу с реализацией одного из изученных алгоритмов решения СЛАУ.__

```python
import numpy as np

def solve_triang(A, b):
    n = np.shape(A)[0]
    x = np.full((n), np.nan)
    for i in range(0, n):
        x_prev = 0
        for j in range(0, i):
            x_prev = x_prev + A[i][j]*x[j]
        x[i] = (b[i] - x_prev) / A[i][i]
    return x

def LU_forward(A):
    n = np.shape(A)[0]
    U = np.array(A, dtype='float')
    L = np.eye(n)
    for j in range(0, n-1):
        for i in range(j+1, n):
            L[i][j] = U[i][j]/U[j][j]
            U[i][:n] = U[i][:n] - U[j][:n] * L[i][j]
    return  L, U

def LU_backward(L, U, b):
    n = np.shape(L)[0]
    y = solve_triang(L, b)
    x = solve_triang(np.rot90(U, 2), np.flip(y))
    return x
        
def LU(A, b):
    try:
        L, U = LU_forward(A)
        X = LU_backward(L, U, b)
        print(f'L=\n{L}\n')
        print(f'U=\n{U}\n')
        print(f'X={X}\n')
        return  X
    except Exception as e:
        print('wrong dimensions!')
        print(e)
```
